```python
import cv2
import matplotlib.pyplot as plt
import os
import random
import numpy as np
```

```python
image_folder = '/content/drive/MyDrive/BSR/BSDS500/data/images/test/'
```

```python
image_files = [f for f in os.listdir(image_folder) if f.endswith('.jpg')]
print(f"Found {len(image_files)} images.")
```

```python
import random
sample_files = random.sample(image_files, 5)
```

```python
for img_name in sample_files:
    img_path = os.path.join(image_folder, img_name)

    img = cv2.imread(img_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Apply light Gaussian blur (same as before)
    blur = cv2.GaussianBlur(gray, (5, 5), 0)

    # --- Binary Inverse Thresholding (replace Adaptive here) ---
    # You can adjust threshold value manually
    THRESH_VAL = 127  # adjust between 0â€“255
    MAX_VAL = 255

    _, binary_inv = cv2.threshold(
        blur, THRESH_VAL, MAX_VAL, cv2.THRESH_BINARY_INV
    )

    # --- Plot results ---
    plt.figure(figsize=(10, 4))
    plt.suptitle(f"Binary Inverse Thresholding on {img_name}", fontsize=14)

    plt.subplot(1, 2, 1)
    plt.imshow(gray, cmap='gray')
    plt.title('Original Grayscale')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(binary_inv, cmap='gray')
    plt.title('Binary Inverse Thresholded')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
