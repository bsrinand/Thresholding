```python
import cv2
import matplotlib.pyplot as plt
import os
import random
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
image_folder = '/content/drive/MyDrive/BSR/BSDS500/data/images/test/'
```

```python
image_files = [f for f in os.listdir(image_folder) if f.endswith('.jpg')]
print(f"Found {len(image_files)} images.")
```

```python
import random
sample_files = random.sample(image_files, 5)
```

```python
for img_name in sample_files:
    img_path = os.path.join(image_folder, img_name)

    img = cv2.imread(img_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    blur = cv2.GaussianBlur(gray, (5, 5), 0)
    _, otsu = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    plt.figure(figsize=(10, 4))
    plt.suptitle(f"Otsu's Thresholding on {img_name}", fontsize=14)

    plt.subplot(1, 2, 1)
    plt.imshow(gray, cmap='gray')
    plt.title('Original Grayscale')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(otsu, cmap='gray')
    plt.title("Otsu's Thresholded")
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
